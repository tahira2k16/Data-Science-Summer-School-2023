```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

# Independent random variables
X_independent = np.random.randn(1000)
Y_independent = np.random.randn(1000)

# Dependent random variables (linear relationship)
X_dependent = np.random.randn(1000)
Y_dependent = 2 * X_dependent + np.random.randn(1000) * 0.5
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_independent, Y_independent, alpha=0.6)
plt.title('Independent Random Variables')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 2, 2)
plt.scatter(X_dependent, Y_dependent, alpha=0.6)
plt.title('Dependent Random Variables')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

```python
cov_matrix_independent = np.cov(X_independent, Y_independent)
cov_matrix_dependent = np.cov(X_dependent, Y_dependent)

print("Covariance matrix for Independent Variables:\n", cov_matrix_independent)
print("\nCovariance matrix for Dependent Variables:\n", cov_matrix_dependent)
```
