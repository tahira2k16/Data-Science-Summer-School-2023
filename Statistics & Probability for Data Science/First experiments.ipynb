```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

## Flipping a coin

```python
def flip_coin(n, p):
    result = []
    for _ in range(n):
        result.append('H' if random.random() < p else 'T')
    return result

def plot_histogram(data, n):
    fig, axs = plt.subplots(2)

    # Frequency
    axs[0].hist(data, bins=[0,1,2], align='left', rwidth=0.8, color='blue', alpha=0.7)
    axs[0].set_xlabel('Coin Value')
    axs[0].set_ylabel('Frequency')
    axs[0].set_title('Coin Flip Distribution - Frequency')

    # Relative Frequency
    axs[1].hist(data, bins=[0,1,2], align='left', rwidth=0.8, color='green', alpha=0.7, weights=[1./n]*len(data))
    axs[1].set_xlabel('Coin Value')
    axs[1].set_ylabel('Relative Frequency')
    axs[1].set_title('Coin Flip Distribution - Relative Frequency')

    # Adjust layout for better visualization
    plt.tight_layout()
    plt.show()

N = 5  # Number of times you want to flip the coin
p = 0.5  # The probability of getting heads
flip_results = flip_coin(N, p)
plot_histogram(flip_results, N)
plt.show()
```

# Rolling a dice

```python
import random
import matplotlib.pyplot as plt

def roll_dice(n):
    result = []
    for _ in range(n):
        result.append(random.randint(1,6))
    return result

def plot_histogram(data, n):
    fig, axs = plt.subplots(2)
    
    # Frequency
    axs[0].hist(data, bins=range(1,8), align='left', rwidth=0.8, color='blue', alpha=0.7)
    axs[0].set_xlabel('Dice Value')
    axs[0].set_ylabel('Frequency')
    axs[0].set_title('Dice Roll Distribution - Frequency')

    # Relative Frequency
    axs[1].hist(data, bins=range(1,8), align='left', rwidth=0.8, color='green', alpha=0.7, weights=[1./n]*len(data))
    axs[1].set_xlabel('Dice Value')
    axs[1].set_ylabel('Relative Frequency')
    axs[1].set_title('Dice Roll Distribution - Relative Frequency')

    # Adjust layout for better visualization
    plt.tight_layout()
    plt.show()

N = 100000  # Number of times you want to roll the dice
roll_results = roll_dice(N)
plot_histogram(roll_results, N)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_normal_samples_histogram(n_samples, mean, variance):
    # Draw samples from a normal distribution
    samples = np.random.normal(loc=mean, scale=np.sqrt(variance), size=n_samples)
    
    # Create the histogram
    plt.hist(samples, bins=30, edgecolor='black')

    # Set the title and labels
    plt.title(f"Histogram of {n_samples} samples from a normal distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Show the plot
    plt.show()

# Test the function
draw_normal_samples_histogram(1000000, 0, 1)
```
