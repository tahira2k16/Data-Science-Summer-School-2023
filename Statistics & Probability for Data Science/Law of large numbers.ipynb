```python
import numpy as np
import matplotlib.pyplot as plt

def coin_toss(n, p=0.5):
    """Simulate n coin tosses with probability p of getting a heads."""
    return np.random.binomial(1, p, n)

def plot_empirical_mean(n, p=0.5):
    """Plot the empirical mean of n coin tosses over time."""
    tosses = coin_toss(n, p)
    empirical_means = np.cumsum(tosses) / np.arange(1, n + 1)

    plt.figure(figsize=(10, 6))
    plt.plot(np.arange(1, n + 1), empirical_means, label=f'Empirical mean (p={p})')
    plt.axhline(y=p, color='r', linestyle='--', label=f'Expected value (p={p})')
    plt.xlabel('Number of Tosses')
    plt.ylabel('Empirical Mean')
    plt.title('Law of Large Numbers - Coin Tosses')
    plt.legend()
    plt.grid(True)
    plt.show()

# Number of tosses
n = 1000

# Probability of heads (try changing this to see behavior for different biases)
p = 0.8

plot_empirical_mean(n, p)
```
