```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set a random seed for reproducibility
np.random.seed(0)

# Generate samples from a normal distribution with mean=2 and standard deviation=1
data = np.random.normal(0, 1, 1000)

# Generate Q-Q plot comparing the above data to a standard normal distribution
fig = plt.figure()
ax = fig.add_subplot(111)
stats.probplot(data, dist="norm", plot=ax)

ax.set_title("Q-Q plot: Comparing data to a standard normal distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set a random seed for reproducibility
np.random.seed(0)

# Generate samples
normal_data = np.random.normal(0, 1, 1000)
cauchy_data = np.random.standard_cauchy(1000)
t_data = np.random.standard_t(2, 1000)

# Create Q-Q plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Q-Q plot comparing normal data to standard normal distribution
stats.probplot(normal_data, dist="norm", plot=axes[0])
axes[0].set_title("Q-Q plot: Normal vs. Standard Normal")

# Q-Q plot comparing Cauchy data to standard normal distribution
stats.probplot(cauchy_data, dist="norm", plot=axes[1])
axes[1].set_title("Q-Q plot: Cauchy vs. Standard Normal")

plt.tight_layout()
plt.show()

# Q-Q plot comparing t data to standard normal distribution
fig, ax = plt.subplots(figsize=(6, 5))
stats.probplot(t_data, dist="norm", plot=ax)
ax.set_title("Q-Q plot: t (df=2) vs. Standard Normal")
plt.show()
```
