## Task 3: Derivatives

Implement derivative calculator and calculate the slope at $2$ for function $f(x) = x^2$

```python
def derivative_x(f, x, step_size):
  #TODO
  m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
  return m

def my_function(x):
  return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2)
```

```python
# Chain Rule y=x^2+1, z=y^3-2

x, y = symbols('x y')

# derivative for first function
# need to underscore y to prevent variable clash
_y = x**2 + 1
dy_dx = diff(_y)

# derivative for second function
z = y**3 - 2
dz_dy = diff(z)

# Calculate derivative with and without
# chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

# Prove chain rule by showing both are equal
print(dz_dx_chain) # 6*x*(x**2 + 1)**2
print(dz_dx_no_chain) # 6*x*(x**2 + 1)**2
```

```python
# Integral approximation

# Implement approximate integral function
def approximate_integral(a, b, n, f):
  #TODO
  delta_x = (b-a)/n
  total_sum = 0

  for i in range(1, n+1):
    midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
    total_sum += f(midpoint)

  return total_sum * delta_x

def my_function(x):
  return x**2 + 1

area = approximate_integral(a=2, b=3, n=5, f=my_function)

print(area) # prints 7.330000000000002
```

## $$\textbf{ Multivariate function derivatives}$$
<hr>
<br>
<center>
$y = 3 x_1^3 + 2^{x_2}, \quad \frac{\partial y}{\partial x_1} = 9 x_1^2  \quad  \frac{ \partial y}{\partial x_2} = 2^{x_2} \ln{2}$

```python
x1 = 2. # has to be a float number
x2 = 3. # has to be a float number

# Implement y
#TODO
y = lambda x1, x2 : 3 * np.power(x1, 3) + np.power(2, x2)
```

```python
# wrt x_1 (first func arg)
val = grad(y, 0)(x1, x2) # 9 * 2**2 = 36.

test_val = 9*x1**2
print(f'{val} == {test_val}')

assert np.isclose(val,test_val)
```

```python
# wrt x_2
val = grad(y, 1)(x1, x2) # 2**(3)*ln(2)

test_val = 2**x2 * np.log(2)
print(f'{val} == {test_val}')

assert np.isclose(val, test_val, atol=1e-3) # 0.001 10**-3
```

## $$\textbf{ Gradient descent for finding minimum of a function}$$
<hr>
<br>
$$y = kx^2 + const, \quad k=3$$

```python
# a function to find a minimum numerically
import jax.numpy as np

f = lambda x, coeff=3, offset=0: coeff * np.power(x, 2) + offset # jax numpy !

def minimize_(fn, x0, step_size, max_it):

    vals = np.array([x0]).astype(np.float32) # list of the checked points

    for _ in range(max_it):

        grad_f = grad(fn)

        curr_val = vals[-1]
        if abs(grad_f(curr_val)) > 1e-5: # the derivative approx not zero

            #TODO
            next_val = curr_val - step_size*grad_f(curr_val)
            vals = np.append(vals, next_val)

        else:
            return vals

    return vals
```

```python
vals = minimize_(f, x0=-5., step_size=0.1, max_it=10)
vals
```

## $$\textbf{ Plot the optimization steps}$$
<hr>

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_minimize(x, f, max_it=10, step_size=0.01):

    plt.figure(figsize=(10, 8))
    plt.plot(x, f(x), label='$f(x)$')

    vals = minimize_(f, x0=x[0], step_size=step_size, max_it=max_it)

    plt.plot(vals, f(vals), "o--", c="r", label='opt step')
    plt.legend(loc='best', fontsize=20)
    plt.grid()
```

```python
x = np.arange(-5, 5.01, 0.25) # function range
plot_minimize(x, f, max_it=10, step_size=0.1)
```

# Object-Oriented Implementation
We would like to make use of existing neural network objects of the PyTorch library. Here, most of the code is already implemented for you. You are only asked to compute the forward pass and the error gradient.

```python
import torch
import torch.nn as nn

# 1. Get the data and parameters

X, T = getdata()
W, B = getparams()

# 2. Convert to PyTorch objects

X = torch.Tensor(X)
T = torch.Tensor(T)
W = [torch.nn.Parameter(torch.Tensor(w.T)) for w in W]
B = [torch.nn.Parameter(torch.Tensor(b)) for b in B]

# 3. Build the neural network

net = torch.nn.Sequential(
            nn.Linear(4,6),nn.ReLU(),
            nn.Linear(6,6),nn.ReLU(),
            nn.Linear(6,6),nn.ReLU(),
            nn.Linear(6,1))

for l,w,b in zip(list(net)[::2],W,B):
    l.weight = w
    l.bias = b

# 4. Compute the forward pass and the error gradient

#TODO
Y = net.forward(X)
err = ((Y-T)**2).mean()
err.backward()


# 5. Show error gradient w.r.t. the 1st weight parameter
print(numpy.linalg.norm(list(net)[0].weight.grad[0,0]))
```
